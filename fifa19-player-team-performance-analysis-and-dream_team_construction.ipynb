```python
#Height Column
def clean_height(x):
    l=[]
    s=1
    if isinstance(x,str):
        l=x.split("'")
        i=int(l[0])
        j=int(l[1])
        s=((i*12)+j)/12
    return(s)


fifa['Height'] = fifa['Height'].apply(clean_height)
fifa['Height'].fillna((fifa['Height'].mean()), inplace = True)
```

```python
#Skill Columns

#Function to convert skill rating at each position.
def skillConverter(val):
    if type(val) == str:
        s1 = val[0:2]
        s2 = val[-1]
        val = int(s1) + int(s2)
        return val
    else:
        return val

skill_columns = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']

for col in skill_columns:
    fifa[col] = fifa[col].apply(skillConverter)
    fifa[col].fillna(0.0, inplace = True)
```

```python
import pandas as pd
import plotly.graph_objects as go

def pie_count(data, field="Nationality", percent_limit=0.5):
    title = f"Number of players by {field}"
    vc = data[field].value_counts(normalize=True) * 100
    main_data = vc[vc >= percent_limit]
    others = vc[vc < percent_limit].sum()
    labels = main_data.index.tolist() + [f"Others(<{percent_limit}% each)"]
    values = main_data.tolist() + [others]

    fig = go.Figure(data=[go.Pie(labels=labels, values=values)],
                    layout=go.Layout(title=title, height=700))
    fig.show()

# Example usage
pie_count(fifa, 'Nationality')
pie_count(fifa, 'Club')
pie_count(fifa, 'Preferred Foot')
pie_count(fifa, 'Work Rate', 0.1)
pie_count(fifa, 'Body Type', 0.1)
pie_count(fifa, 'Position', 0.1)
```

```python
import pandas as pd

player_features = [
    'Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling', 
    'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 
    'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength', 
    'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 
    'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling', 
    'GKKicking', 'GKPositioning', 'GKReflexes'
]

df_position = pd.DataFrame({
    position: tuple(dict(features.nlargest(5)))
    for position, features in fifa.groupby('Position')[player_features].mean().iterrows()
})

df_position.head()
```

```python

positions, players, clubs = [], [], []

for col in df_position.columns:
    data = fifa[fifa['Position'] == col]
    data['mean'] = data.mean(axis=1)
    best_player = data.loc[data['mean'].idxmax()]
    positions.append(col)
    players.append(best_player['Name'])
    clubs.append(best_player['Club'])

categories = {
    'Goalkeeper': ['GK'],
    'Forward': ['LS', 'ST', 'RS', 'LF', 'CF', 'RF'],
    'Midfielder': ['LW', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LDM', 'CDM', 'RDM'],
    'Defender': ['LWB', 'RWB', 'LB', 'LCB', 'CB']
}

for category, positions_list in categories.items():
    print(f'\n{category.upper()} : ')
    for pos, name, club in zip(positions, players, clubs):
        if pos in positions_list:
            print(f'{name} [Club: {club}, Position: {pos}]')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data setup (ensure this matches your actual setup)
# This is an example and should be replaced with your actual DataFrame loading method
data = {
    'Name': ['Player1', 'Player2', 'Player3', 'Player4', 'Player5'],
    'Club': ['Club1', 'Club2', 'Club3', 'Club4', 'Club5']
}
fifa = pd.DataFrame(data)

def create_football_formation(formation=[], label_1=None, label_2=None, label_3=None, label_4=None, label_4W=None, label_5=None, label_3W=None):
    plt.figure(figsize=(15, 7))
    plt.scatter(x=[1], y=[6], s=300, color='blue')
    plt.annotate('De Gea \n(Manchester United)', (0.5, 6.5))
    
    n = 0
    label_map = {1: label_1, 2: label_2, 3: label_3, 4: label_4, 5: label_5, '3W': label_3W, '4W': label_4W}
    
    for posi in formation:
        if label_map.get(posi):
            n += 3
            players = label_map[posi]
            y = np.linspace(10/(len(players)+1), 10-10/(len(players)+1), len(players))
            x = [n] * len(players)
            
            plt.scatter(x, y, s=400, color='white')
            plt.scatter(x, y, s=300, color='red')
            
            for i, player in enumerate(players):
                club = fifa['Club'][fifa['Name'] == player].values[0] if not fifa[fifa['Name'] == player].empty else 'Unknown Club'
                plt.annotate(f'{player}\n({club})', (x[i]-0.5, y[i]+0.5))
            
            if posi in ['3W', '4W'] and label_map['%sW' % posi]:  # Special case for additional wide players
                n += 3
                players_wide = label_map['%sW' % posi]
                y = np.linspace(10/(len(players_wide)+1), 10-10/(len(players_wide)+1), len(players_wide))
                x = [n] * len(players_wide)
                
                plt.scatter(x, y, s=400, color='white')
                plt.scatter(x, y, s=300, color='red')
                
                for i, player in enumerate(players_wide):
                    club = fifa['Club'][fifa['Name'] == player].values[0] if not fifa[fifa['Name'] == player].empty else 'Unknown Club'
                    plt.annotate(f'{player}\n({club})', (x[i]-0.5, y[i]+0.5))
    
    plt.plot(np.ones((5, ))*(n+0.5), np.arange(4, 9), 'w-')
    plt.plot(np.ones((11, ))*(n/2), np.arange(1, 12), 'w-')
    plt.yticks([])
    plt.xticks([])
    ax = plt.gca()
    ax.set_facecolor('#28fc03')
    plt.title('Football Formation')
    plt.show()

# Example usage:
create_football_formation(formation=[4, 2], label_4=['Player1', 'Player2', 'Player3', 'Player4'], label_2=['Player5'])
```

```python
#Get a count of Nationalities in the Dataset, make of list of those with over 250 Players (Major Playing Nations)
nat_counts = fifa["Nationality"].value_counts()
nat_list = nat_counts[nat_counts > 250].index.tolist()

#Replace Nationality with a binary indicator variable for 'Major Nation'
def major_nation(df):
    if (df["Nationality"] in nat_list):
        return 1
    else:
        return 0
    
fifa['Major_Nation'] = fifa.apply(major_nation,axis = 1)
```

```python
#Turn Preferred Foot into a binary indicator variable
def right_footed(df):
    if (df['Preferred Foot'] == 'Right'):
        return 1
    else:
        return 0
fifa['Right_Foot'] = fifa.apply(right_footed, axis=1)
```

```python
#Create a simplified position varaible to account for all player positions
def simple_position(df):
    if (df['Position'] == 'GK'):
        return 'GK'
    elif ((df['Position'] == 'RB') | (df['Position'] == 'LB') | (df['Position'] == 'CB') | (df['Position'] == 'LCB') | (df['Position'] == 'RCB') | (df['Position'] == 'RWB') | (df['Position'] == 'LWB') ):
        return 'DF'
    elif ((df['Position'] == 'LDM') | (df['Position'] == 'CDM') | (df['Position'] == 'RDM')):
        return 'DM'
    elif ((df['Position'] == 'LM') | (df['Position'] == 'LCM') | (df['Position'] == 'CM') | (df['Position'] == 'RCM') | (df['Position'] == 'RM')):
        return 'MF'
    elif ((df['Position'] == 'LAM') | (df['Position'] == 'CAM') | (df['Position'] == 'RAM') | (df['Position'] == 'LW') | (df['Position'] == 'RW')):
        return 'AM'
    elif ((df['Position'] == 'RS') | (df['Position'] == 'ST') | (df['Position'] == 'LS') | (df['Position'] == 'CF') | (df['Position'] == 'LF') | (df['Position'] == 'RF')):
        return 'ST'
    else:
        return df.Position
    
fifa['Simple_Position'] = fifa.apply(simple_position,axis = 1)
```

```python
#Split the Work Rate Column in two
tempwork = fifa["Work Rate"].str.split("/ ", n = 1, expand = True) 

#Create new column for first work rate
fifa["WorkRate1"]= tempwork[0]   

#Create new column for second work rate
fifa["WorkRate2"]= tempwork[1]
```

```python
# define some handy analysis support function
from sklearn.metrics import mean_squared_error, explained_variance_score, r2_score

def plot_prediction_analysis(y, y_pred, figsize=(10,4), title=''):
    fig, axs = plt.subplots(1, 2, figsize=figsize)
    axs[0].scatter(y, y_pred)
    mn = min(np.min(y), np.min(y_pred))
    mx = max(np.max(y), np.max(y_pred))
    axs[0].plot([mn, mx], [mn, mx], c='red')
    axs[0].set_xlabel('$y$')
    axs[0].set_ylabel('$\hat{y}$')
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    evs = explained_variance_score(y, y_pred)
    r2 = r2_score(y, y_pred)
    axs[0].set_title('rmse = {:.2f}, evs = {:.2f}, r2 = {:,.2f}'.format(rmse, evs, r2))
    
    axs[1].hist(y-y_pred, bins=50)
    avg = np.mean(y-y_pred)
    std = np.std(y-y_pred)
    axs[1].set_xlabel('$y - \hat{y}$')
    axs[1].set_title('Histrogram prediction error, $\mu$ = {:.2f}, $\sigma$ = {:.2f}'.format(avg, std))
    
    if title!='':
        fig.suptitle(title)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

#model = LinearRegression()
model = Pipeline((
        ("standard_scaler", StandardScaler()),
        ("poly", PolynomialFeatures(degree=2)),
        ("lin_reg", Lasso(alpha=0.01))
    ))

model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)

plot_prediction_analysis(y_train, y_train_pred, title='Polynomial Model - Trainingset')

y_test_pred = model.predict(X_test)
plot_prediction_analysis(y_test, y_test_pred, title='Polynomial Model - Testset')
```
